```python
X=data[['total_rooms','total_bedrooms','population','households','median_income']]
print(X)
print(type(X))
print(X.shape)

y=data['median_house_value']
print(y)
print(type(y))
print(y.shape)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,random_state=0,test_size=0.25)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
train_data=X_train.join(y_train)
train_data
```

```python
trainDataExcel = pd.concat([X_train,y_train],axis=1)
testDataExcel = pd.concat([X_test,y_test],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
import math
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
math.sqrt(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred) #接近0代表模型不好，反之接近1代表好
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
mlr = smf.ols(formula='median_house_value ~  total_rooms + total_bedrooms + population + households + median_income', data=trainDataExcel).fit()
print(mlr.summary())
```

```python
linreg=linear_model.LinearRegression()
linreg.fit(X_train,y_train)
print(linreg.intercept_)
print(linreg.coef_)
```

```python
y_pred=linreg.predict(X_test)
print(y_pred)
```

```python
# create a list to save other_variable_values
other_variable_values_list = []
income_variable_values_list = []

# find the test data index by predicted index
matching_indices = np.arange(len(X_test))[~np.isnan(y_pred)]

if len(matching_indices) > 0:
    print("Matching predictions found at test data positions:")
    for idx in matching_indices:
        test_data_position = idx
        x_value = X_test[test_data_position][0]
        # find other variables index
        other_variable_values = data['postal_code'][test_data_position]
        other_variable_values_list.append(other_variable_values)
        income_variable_values = data['median_income'][test_data_position]
        income_variable_values_list.append(income_variable_values)
        #print(f"Test data position = {test_data_position}, X = {x_value}, Other variables = {other_variable_values}")

else:
    print("No matching predictions found.")

print("Other variable values:", other_variable_values_list)
print("income variable values:", income_variable_values_list)
```

```python
# Create a DataFrame containing other_variable_values_list
df = pd.DataFrame({'postalcode': other_variable_values_list,'Predicted': y_pred,'actual': y_test, 'median_income': income_variable_values_list})

# output to CSV
df.to_csv('postalcode_predicted_actual_variable.csv', index=False)

print("Other variable values have been written to postalcode_predicted_actual_variable.csv")
```
